```python
from helpers.utils import DownloadFromWeights

dl = DownloadFromWeights("output")

blob_list = dl.list_blobs()
```

```python
num_to_download = 1000
csv_store = 'data/json'
image_store = 'data/images'
```

```python
#Download a large number of csv's
csv_list = []
i = 0
for l in blob_list:
    if i > num_to_download:
        break
    if l.name[-4:]=='json':
        csv_list.append(l.name)
        i+=1
```

```python
#dl.download_images(csv_list, csv_store)
import os
for file in csv_list:
    blob_client = dl.blob_service_client.get_blob_client(
    container=dl.blob_container_name, blob=file
    )
    with open(os.path.join(csv_store, file), "wb") as download_file:
        download_file.write(blob_client.download_blob().readall())
```

```python
print(len(csv_list))
```

```python
import json

image_list = []
for f in csv_list:

    with open('data/json/'+f, 'r', encoding='utf-8') as file:
        data = json.load(file)

    for post in data:
        image_url = post['instagram']['image_url']
        image_list.append(image_url)
```

```python
print(len(image_list))
```

```python
dl2 = DownloadFromWeights("instagram")

for file in image_list:
    blob_client = dl2.blob_service_client.get_blob_client(
    container=dl2.blob_container_name, blob=file
    )
    with open(os.path.join(image_store, file), "wb") as download_file:
        download_file.write(blob_client.download_blob().readall())
```
